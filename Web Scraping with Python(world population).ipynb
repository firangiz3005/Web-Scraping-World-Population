```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url = 'https://worldpopulationreview.com/world-cities'
```

```python
page=requests.get(url)
```

```python
soup= BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
table=soup.find_all('table')[1]
```

```python
print(table)
```

```python
columns=table.find_all('th')
```

```python
print(columns)
```

```python
title_data = [data.text.strip() for data in columns]
print(title_data)
```

```python
import pandas as pd
```

```python
column_data = pd.DataFrame(columns=title_data)
```

```python
column_data
```

```python
row_data=table.find_all('tr')
```

```python
for row in row_data[1:]:
    table_data=row.find_all('td')
    all_rows=[rows.text.strip() for rows in table_data]
    length=len(column_data)
    column_data.loc[length]=all_rows
  
```

```python
column_data
```

```python
column_data.to_csv(r'Downloads\cities.csv')
```
